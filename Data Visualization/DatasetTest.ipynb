```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("/content/country_vaccinations.csv.zip")
data
```

```python
# Fill NaN values with 0
data.fillna(0, inplace=True)

# Check for null values
print(data.isnull().sum())

# Check for duplicate rows
print(data.duplicated().sum())

# Remove duplicate rows
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
# Select the columns for plotting
x = data['date']
y = data['total_vaccinations']

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.title('Total Vaccinations Over Time')

# Show the plot
plt.show()
```

```python
# Select the columns for plotting
x = data['date']
y = data['total_vaccinations']

# Create the plot using seaborn
sns.lineplot(x=x, y=y)

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.title('Total Vaccinations Over Time')

# Show the plot
plt.show()
```

```python
# Pie chart
country_counts = data['country'].value_counts().head(15)  # Get top 5 countries
plt.figure(figsize=(12, 12))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 15 Countries with Most Data Entries')
plt.show()
```

```python
# Group data by country and sum total vaccinations
country_vaccinations = data.groupby('country')['total_vaccinations'].sum()

# Create the bar plot
plt.figure(figsize=(150, 50))
country_vaccinations.plot(kind='bar')

# Add labels and title
plt.xlabel('Country')
plt.ylabel('Total Vaccinations')
plt.title('Total Vaccinations by Country')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
# Check if all values in a specific column are non-zero
all_total_vaccinations_non_zero = np.all(data['total_vaccinations'] != 0)
print("All total vaccinations are non-zero:", all_total_vaccinations_non_zero)

# Check if all values in a specific column are greater than a threshold
all_total_vaccinations_above_threshold = np.all(data['total_vaccinations'] > 1000000)
print("All total vaccinations are above 1 million:", all_total_vaccinations_above_threshold)
```

```python
# Calculate the total vaccinations for each country on a specific date
total_vaccinations_on_date = data[data['date'] == '2021-03-10'].groupby('country')['total_vaccinations'].sum()
print(total_vaccinations_on_date)
```
