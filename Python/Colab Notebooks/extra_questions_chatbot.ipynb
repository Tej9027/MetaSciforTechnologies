```python
# Define a dictionary of predefined questions and answers
knowledge_base = {
    "what is your name?": "I am the Chatbot created by Tejveer Singh.",
    "how are you?": "I'm just a program, but I'm doing great! How about you?",
    "what is python?": "Python is a high-level, interpreted programming language known for its readability and simplicity.",
    "what is ai?": "AI stands for Artificial Intelligence, which is the simulation of human intelligence in machines.",
    "what is your purpose?": "I am here to assist you with any questions you may have and provide information."
}

# Variables to keep track of correct and wrong answers
correct_answers = 0
wrong_answers = 0

# Main function to start the chatbot
def start_chatbot():
    print("Welcome to the Enhanced Chatbot!")
    print("Type 'exit' to end the conversation.")

    while True:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            print("Chatbot: Goodbye!")
            print(f"Correct Answers: {correct_answers}")
            print(f"Wrong Answers: {wrong_answers}")
            break
        response, is_correct = get_response(user_input)
        print(f"Chatbot: {response}")
        update_score(is_correct)

# Function to get a response based on user input
def get_response(user_input):
    user_input = user_input.lower()

    if 'hello' in user_input or 'hi' in user_input:
        return greet(), None
    elif 'how are you' in user_input:
        return how_are_you(), None
    elif user_input in knowledge_base:
        return knowledge_base[user_input], True
    else:
        return "I'm sorry, I don't understand that. Can you please rephrase?", False

# Function to handle greeting
def greet():
    return "Hello! How can I help you today?"

# Function to handle 'how are you' response
def how_are_you():
    return "I'm just a program, but I'm doing great! How about you?"

# Function to update the score based on the correctness of the answer
def update_score(is_correct):
    global correct_answers, wrong_answers
    if is_correct is not None:
        if is_correct:
            correct_answers += 1
        else:
            wrong_answers += 1

# Run the chatbot
if __name__ == "__main__":
    start_chatbot()
```
