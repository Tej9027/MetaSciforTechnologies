```python
# for loop in while loop with break
while True:
  num = int(input("enter a number "))
  if num > 0:
    print("positive")
    break
  elif num<0:
    print("negative")
    break
  else:
    print("zero")
    break
```

```python
# break keyword for sstop at that part

for i in range(1,10):
    if i == 5:
        break
    print(i)
```

```python
# continue statement  for skip some part

for i in range(1,10):
    if i == 5:
        continue
    print(i)
```

```python
n = int(input("enter a number: "))
for i in range(2,n): # we use only n because it is divisible by itself
    if n % i == 0:
        print("Not Prime")
        break
    else:
        print("Prime")
```
