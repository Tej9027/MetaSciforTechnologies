Ayman wants to modify one of his text files. In his text file, he wants to start a new line whenever he encounters a number followed by a period(.). So he decides to do the same with the help of the python program. Write a program for Ayman.

```python
file = open("input.txt", "w")
file.write("This is a test.1. Here is a new sentence.2. And another one.3. Let's see how it works.4. as user wants to start a new line whenever he encounters a number followed by a period(.).")
file.close()

file = open("input.txt", "r")
text = file.read()
file.close()

modified_text = ""
i = 0
while i < len(text):
    if i < len(text) - 1 and text[i].isdigit() and text[i + 1] == '.':
        modified_text += '\n' + text[i] + text[i + 1]
        i += 2
    else:
        modified_text += text[i]
        i += 1

file = open("input.txt", "w")
file.write(modified_text)
file.close()
print(modified_text)
```

Arunima created a text file on programming where she wrote the word language incorrectly many times. Now she wants to replace the incorrect word with the correct word. Doing it manually can take a lot of time plus she can miss out few words at the end. To help Arunima to replace the words write a python program. Use the concept of file handling and make changes in the text file.

```python
file = open("programming.txt", "w")
file.write("Programming langauge is essential for computer sciense. Learning a programming langauge can be challenging but rewarding. The langauge Python is widely used in many industreis.")
file.close()

replacements = {
    "langauge": "language",
    "sciense": "science",
    "industreis": "industries"
}

file = open("programming.txt", "r")
text = file.read()
file.close()

for incorrect, correct in replacements.items():
    text = text.replace(incorrect, correct)

file = open("programming.txt", "w")
file.write(text)
file.close()

print(text)
```

Arvind has a huge text file and he wants to put a serial number at the beginning of each line. Doing this task manually can take a lot of time. He is not aware of python file handling. Write a python program that can put a serial number in front of each line in any specified text file.

```python
file = open("huge_file.txt", "w")
file.write("This is the first line.\nThis is the second line.\nThis is the third line.\n")
file.close()

file = open("huge_file.txt", "r")
lines = file.readlines()
file.close()

serial_number = []
for i, line in enumerate(lines, 1):
    serial_number.append(f"{i}. {line}")

file = open("huge_file.txt", "w")
file.writelines(serial_number)
file.close()

for line in serial_number:
    print(line, end='')
```

Shubham had created a BMI calculator to check his health condition daily. But he is not able to record the values. He wants to see how his BMI is changing. So he decided to connect his BMI program with a text file using the concept of file handling. Now, when he uses his BMI program, his height, weight, BMI, and date get added to a text file "bmi.txt". Create a program that can do the same thing for you.


```python
import datetime

def calculate_bmi(weight, height):
    bmi = weight / (height * height)
    return bmi

def record_bmi(weight, height):
    bmi = calculate_bmi(weight, height)
    date = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    with open("bmi.txt", "a") as file:
        file.write(f"{date}, {height}, {weight}, {bmi:.2f}\n")

    print(f"Recorded: Date: {date}, Height: {height}, Weight: {weight}, BMI: {bmi:.2f}")
height = float(input("Enter your height in meters: "))
weight = float(input("Enter your weight in kilograms: "))
record_bmi(weight, height)
```

Himanshi has a text file and she wants to check whether that file contains any numerical value. So to help Himanshi write a python program that can check whether the text file includes any numerical value or not. As text files contain so many lines, so print the line number that has numerical values


```python
file = open("sample_text.txt", "w")
file.write("This is the first line.\nThis is the second line with number 42.\nThis is the third line.\nThe fourth line has 123 and 456.\n")
file.close()

def check_for_numbers_in_file(filename):
    with open(filename, "r") as file:
        lines = file.readlines()

    lines_with_numbers = []
    for i, line in enumerate(lines, 1):
        if any(char.isdigit() for char in line):
            lines_with_numbers.append(i)

    if lines_with_numbers:
        print(f"The file contains numerical values in the following lines: {lines_with_numbers}")
    else:
        print("The file does not contain any numerical values.")

check_for_numbers_in_file("sample_text.txt")
```
