```python
# Map Function
def task(num):
  return num%2

numbers = [2,3,6,4,3,7,8]

map1 = map(task, numbers)

print(list(map1))
```

```python
# Filter Function filtering out the data on your condition
def task(num):
  return num%2 == 0

numbers = [2,3,6,4,3,7,8]

map1 = filter(task, numbers)

print(list(map1))
```

```python
# lambda function is an anonymas fucntion you don't need a name
#without name you can perfrom task with lambda funtion (you don'thave to use def keyword)

Addition = lambda a,b : a+b

print(Addition(2,3))
```

```python
# list comprehensions

res = [i**2 for i in range(10)]
print(res)
```

```python
l = [(i*i/2) for i in range(10)]
print(l)
```

```python
# Generators used yield fucntion. A Generator in Python is a function that returns an iterator using the Yield keyword.
# yield means return
def generator():
  yield 1
  yield 2
  yield 3

for i in generator():
  print(i)
```
