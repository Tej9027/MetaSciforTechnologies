```python
bank_balance = 50000
print("ATM Menu")
print("1. Withdraw")
print("2. Depost")
print("3. Balance Check")
print("4. Fast Cash")
choice = int(input("enter your choice"))
if choice == 1:
  amount = int(input("enter amount to withdraw(multiple of 100)"))
  if amount % 100 != 0:
    print("please enter the amount in multiple of 100 like 100,200,1000,2000.")
  else:
    bank_balance -= amount
    print(f"You have withdrawn {amount}. New Balance is now {bank_balance}")
elif choice == 2:
  amount = int(input("enter amount to deposit)"))
  bank_balance += amount
  print(f"You have deposited {amount}. Now New bank balance is {bank_balance}")
elif choice == 3:
  print(f"your current balance is {bank_balance}")
elif choice == 4:
  print("Fast cash Options")
  print("1. 5000")
  print("2. 10000")
  amount = int(input("Enter amount from options"))
  if amount == 1:
    bank_balance -= amount
    print(f"you have withdrawn {amount}. so new balance{bank_balance}")
  else:
    bank_balance -= amount
    print(f"you have withdrawn {amount}. so new balance{bank_balance}")
else:
  print("Invalid Options. Please chosse correct options.")
```
