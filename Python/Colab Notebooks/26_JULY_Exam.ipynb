Create a student management system project using the classes and objects concept in python which should perform the following operations

1.Accept Student details

2.Display Student Details

3.Search Details of a Student

4.Delete Details of Student

5.Update Student Details

6.Exit

#create a student class and add functions for the above shown options in the description and ask the user to choose one among them

```python
class Student:
  def __init__(self,name,roll_no,age,course,marks):
    self.name=name
    self.roll_no=roll_no
    self.marks=marks
    self.age=age
    self.course=course
  def display_details(self):
    print(f"Name: {self.name}")
    print(f"Roll No: {self.roll_no}")
    print(f"Age: {self.age}")
    print(f"Course: {self.course}")
    print(f"Marks: {self.marks}")

class StudentManagementSystem:
  def __init__(self):
    self.students=[]

  def accept_student_details(self):
    name=input("Enter name: ")
    roll_no=input("Enter roll no: ")
    age=input("Enter age: ")
    course=input("Enter course: ")
    marks=input("Enter marks: ")
    student = Student(name,roll_no,age,course,marks)
    self.students.append(student)
    print("Details of student added successfully")

  def display_student_details(self):
    if not self.students:
      print("No student details found")
    else:
      for student in self.students:
        student.display_details()
        print()

  def search_student_details(self):
    roll_no = int(input("Enter roll no to search: "))
    for student in self.students:
      if student.roll_no == roll_no:
        student.display_details()
        return
    print("Student details not found")


  def delete_student_details(self):
    roll_no = int(input("Enter roll no to delete: "))
    for student in self.students:
      if student.roll_no == roll_no:
        self.students.remove(student)
        print("Student deleted successfully")
        return
    print("Student details not found")

  def update_student_details(self):
    roll_no = int(input("Enter roll no to update: "))
    for student in self.students:
      if student.roll_no == roll_no:
        student.name = input("Enter new name: ")
        student.age = int(input("Enter new age: "))
        student.course = input("Enter new course: ")
        student.marks = input("Enter new marks: ")
        print("Student details updated successfully")
        return
    print("Student details not found")

def main():
  print("Welcome to Student Management System")
  sms = StudentManagementSystem()
  while True:
    print("1. Accept Student Details")
    print("2. Display Student Details")
    print("3. Search Student Details")
    print("4. Delete Student Details")
    print("5. Update Student Details")
    print("6. Exit")
    choice = int(input("Enter your choice: "))
    if choice == 1:
      sms.accept_student_details()
    elif choice == 2:
      sms.display_student_details()
    elif choice == 3:
      sms.search_student_details()
    elif choice == 4:
      sms.delete_student_details()
    elif choice == 5:
      sms.update_student_details()
    else:
      print("wrong input")

main()
```
