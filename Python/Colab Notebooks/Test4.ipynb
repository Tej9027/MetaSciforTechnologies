```python
# A Candy Vending Machine is topped up and is supposed to deliver the number of candies entered by the user
# but before delivering the candies, the vending machine is supposed to ask the user to enter the amount for the candies
# where each candy is worth 1 dollar.

# Step 1: Load the vending machine with candies
total_candies_to_load = int(input("How many candies to be loaded: "))

# Step 2: Display the vending machine title and number of loaded candies
print(f"\nüç¨üç¨ Candy Vending Machine üç¨üç¨")
print(f"{total_candies_to_load} candies loaded into the vending machine\n")

# Step 3: Accept the number of candies the customer wants
numberOfCandiesUserWants = int(input("Enter the number of candies you want: "))

# Step 4: Validate the number of candies requested
if numberOfCandiesUserWants <= 0 or numberOfCandiesUserWants > total_candies_to_load:
    print("\nYou have entered a number of candies which is either zero, less than zero, or more than available.")
    print("Please try again or visit once again. Thanks for shopping with us! We will be happy to serve you once again.")
else:
    # Step 5: Calculate the total cost and display it
    cost = numberOfCandiesUserWants
    print(f"\nYou have entered {numberOfCandiesUserWants} candies and you have to pay ${cost}.\n")

    # Step 6: Confirm the money paid by the customer
    UserMoney = int(input("Please enter the amount you are paying: "))

    # Step 7: Validate the payment
    if UserMoney == cost:
        print("\nDelivering candies...")

        # Delivering candies message without using loops
        print(f"Delivered {numberOfCandiesUserWants} candies")

        # Step 8: Deduct the delivered candies from the total
        total_candies_to_load -= numberOfCandiesUserWants
        print(f"\nThe number of candies left in the vending machine are {total_candies_to_load}")
        print("Thank you for shopping with us!")
    else:
        print("\nEntered wrong amount. Try again!")
        print("Thank you for shopping with us!")
```

```python
start = int(input("Enter the starting range (greater than 10): "))
end = int(input("Enter the ending range: "))

if start <= 10:
    print("Starting range should be greater than 10.")
else:
    print("Even numbers in the range:")
    for number in range(start, end + 1):
        if number % 2 == 0:
            print(number, end=" ")

    print("\nOdd numbers in the range:")
    for number in range(start, end + 1):
        if number % 2 != 0:
            print(number, end=" ")
```

```python
# Get input from the user
input_string = input("Enter a string: ")

# Initialize counters
upper_count = 0
lower_count = 0
digit_count = 0
symbol_count = 0

# Loop through each character in the string
for char in input_string:
    if char.isupper():
        upper_count += 1
    elif char.islower():
        lower_count += 1
    elif char.isdigit():
        digit_count += 1
    else:
        symbol_count += 1

# Print the results
print(f"Total Uppercase Letters: {upper_count}")
print(f"Total Lowercase Letters: {lower_count}")
print(f"Total Digits: {digit_count}")
print(f"Total Symbols: {symbol_count}")
```

```python
# Get the year from the user
year = int(input("Enter a year: "))

# Check if the year is a leap year
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```
