write a python program of finding prime numbers

```python
def prime_number(n, lower,upper):
  print("Prime numbers between",n,"and",upper,"are:")
  for n in range(lower,upper+1):
    if n > 1:
      for i in range(2,n):
        if( n % i == 0):
          break
      else:
        print(n)
prime_number(2,1,100)
```

write a python program of creating chatbot major one using functions

```python
def chatbot():
    correct = 0
    wrong = 0

    print("Hi! I am Chatbot! I am here to chat and play a quiz with you.")

    # Normal interactions
    print("Hello! How are you today?")
    user_response = input("You: ")

    if user_response.lower() == 'exit':
            print("Chatbot: Goodbye! Have a nice day!")
            return
    if "good" in user_response.lower() or "fine" in user_response.lower():
        print("Chatbot: I'm glad to hear that!")
    elif "not" in user_response.lower() or "bad" in user_response.lower():
        print("Chatbot: I'm sorry to hear that. I hope things get better!")
    else:
        print("Chatbot: Interesting! Let's move on to the quiz.")

    print("Let's start the quiz!")

    # Define a list of questions and their correct answers
    quiz = [
        ("What is 12^2?", 144),
        ("What is 13*3?", 39),
        ("What is 2*2?", 4),
        ("What is 1^1?", 1),
        ("What is the capital of France?", "Paris"),
        ("What is 10 + 15?", 25),
        ("What is the square root of 81?", 9),
        ("What is 7 - 3?", 4),
        ("What is 5 * 6?", 30),
        ("What is 100 / 25?", 4)
    ]

    for question, answer in quiz:
        print(question)
        if isinstance(answer, int):
            user_answer = int(input())
        else:
            user_answer = input()

        if user_answer == answer:
            correct += 1
        else:
            wrong += 1

    print(f"You got {correct} correct answers and {wrong} wrong answers.")

chatbot()
```
