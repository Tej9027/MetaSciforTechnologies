```python
while True:
  balance = 10000
  print("\nWelcome to the ATM\n 1. Check Balance\n 2. Deposit Money\n 3. Withdra Money\n 4. Exit\n")

  choice = input("Choose an option (1-4): ")

  if choice == "1":
    print(f"Your Balance is: ${balance}")

  elif choice == "2":
    amount = float(input("enter amount to depoist: $"))
    if amount > 0:
      balance += amount
      print(f"Deposited Amount is ${amount: }. New Balance Amount is: ${balance: }")
    else:
      print("Invalid amount.")
  elif choice == "3":
    amount = float(input("Enter amount to withdra: $"))
    if 0 < amount <= balance:
      balance -= amount
      print(f"Withdrawl Amount is ${amount}. New Balance Amount is: ${balance}")
    elif amount > balance:
      print("Insufficient Balance.")
    else:
      print("Invalid Amount. Please enetr a valid amount.")
  elif choice == "4":
        # Exit
        print("Thank you for using the ATM. Goodbye!")
        break
  else:
    print("Invalid choice. Please try again.")
```

```python
print("Welcome to SimpleChatBot!")
print("Type 'exit' to end the chat.\n")

while True:
    user_input = input("You: ").strip().lower()

    if user_input == 'exit':
        print("ChatBot: Goodbye! Have a great day!")
        break
    elif user_input in ['hi', 'hello']:
        print("ChatBot: Hello! How can I help you today?")

    elif user_input in ['how are you?', 'how are you']:
        print("ChatBot: I'm just a bot, but I'm doing great! How about you?")

    elif user_input in ['what is your name?']:
        print("ChatBot: I'm SimpleChatBot, your virtual assistant!")

    elif user_input in ['current time?']:
        print("ChatBot: Sorry, I can't provide the time, but you can check your watch or phone!")
    else:
        print("ChatBot: I'm not sure how to respond to that. Try asking something else.")

print("ChatBot session ended.")
```
