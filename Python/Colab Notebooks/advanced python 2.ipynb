```python
# Decorators in Python

#decorators are used for change the functionality of the code.
```

```python
def dec(func): # this parameter we can use this as a funtion
  def wrap():
    print("Free")
    func() #original fucntion
    print("You")
  return wrap

@dec
def decoratorsresult():
  print("Kalpna")

decoratorsresult()
```

```python
def dec(func):
  def wrap():
    print("Free")
    func()
    print("Kalpna")
  return wrap

@dec
def decoratorsresult():
  print("You")

decoratorsresult()
```

```python
# 1st Program
def uppercase_decorator(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper

@uppercase_decorator
def greet():
    return "hello, world!"

print(greet())
```

```python
# 2nd program
def simple_logger(func):
  def wrapper():
    print("function is starting")
    func()
    print("function is finished")
  return wrapper

@simple_logger
def simple_function():
  print("inside the function")

simple_function()
```

```python
# 3rd Program
def repeat(func):
  def wrapper():
    for _ in range(5):
      func()
  return wrapper

@repeat
def function():
  print("kalpna")

function()
```
