Create a program of finding perimeter of a triangle and area of any three shapes using class object

```python
class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def perimeter(self):
        return self.a + self.b + self.c

    def area(self):
        s = (self.a + self.b + self.c) / 2
        return (s * (s - self.a) * (s - self.b) * (s - self.c)) ** 0.5

a = float(input("Enter the length of side a: "))
b = float(input("Enter the length of side b: "))
c = float(input("Enter the length of side c: "))

triangle = Triangle(a, b, c)
print("Perimeter of the triangle:", triangle.perimeter())
print("Area of the triangle:", triangle.area())
```

Print the sum of 1000 numbers using class object concepts

```python
class sum:
    def __init__(self, n):
        self.n = n
        self.sum = 0

    def calculate_sum(self):
        for i in range(1, self.n + 1):
            self.sum += i
        return self.sum

n = int(input("Enter the value of n: "))
s = sum(n)
print("Sum of first", n, "numbers:", s.calculate_sum())
```

Create a factorial program using class object

```python
class factorial:
  def __init__(self, n):
    self.n = n
    self.fact = 1

  def factorial(self):
    for i in range(1, self.n + 1):
      self.fact *= i
    return self.fact

n = int(input("Enter the value of n: "))
f = factorial(n)
print("Factorial of", n, "is:", f.factorial())
```

Create a. Fibonacci series using class object

```python
class fibonacci:
  def __init__(self, n):
    self.n = n
    self.a = 0
    self.b = 1
  def fibonacci(self):
    print(self.a)
    if self.n > 1:
      print(self.b)
    count = 2
    while count < self.n:
      self.a, self.b = self.b, self.a + self.b
      print(self.b)
      count += 1
n = int(input("Enter the value of nth term: "))
f = fibonacci(n)
f.fibonacci()
```
