5 friends decided to go for a picnic and they start discussing the items which they have to take with them. They wanted to create a list where they can add items one by one. To help them, write a program in python to create an empty list. Then use for loop and append() method to add items one by
one. ( Note: Range of loop depends on the number of items that have to be included in the list.) After adding all the items ask whether the user whether wants to remove any items or not and then print the whole list on the screen.

```python
def picnic():
  item = []
  items = int(input("enter the number of items: "))
  for i in range(items):
    item.append(input("enter the item: "))
  print(item)

  delete = input("if you want to delete an item from the list then enter yes else no: ")

  if delete == "yes":
    item.remove(input("enter the item to be deleted: "))
    print(item)
  else:
    print(item)

picnic()
```

 Lastly, she wants to find the mode of numbers. To help Nidhi in this problem, write a program in python where users can enter n numbers as input, and then the program will print the mode of numbers. ● Hint: Print the number having maximum count in the list.

```python
user_input = input("Enter a list of numbers separated by spaces: ")
numbers = [int(num) for num in user_input.split()]
count = {}
for num in numbers:
  if num in count:
    count[num] += 1
  else:
    count[num] = 1
  mode = max(count, key=count.get)
print(f"The mode of the numbers is: {mode}")
print(f"The count of the numbers is: {count}")
```
