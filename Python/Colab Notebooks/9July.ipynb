```python
# function with parameter (is just like variable names)
def add(x,y):
  return x+y

add(3,6)
```

```python
# globar variable used outside
name = 's'
def task():
  print(name)
task()
print(name)
```

```python
#local variable not used outide
def task():
  name2 = 's'
  print(name2)
task()
print(name2)
```

```python
#local variable not used outide
def task():
  global name2
  name2 = 's'
  print(name2)
task()
print(name2)
```
