```python
# GET Request
import requests
response = requests.get('https://api.chucknorris.io/jokes/random')
if response.status_code == 200:
  value = response.json()['value']
  print(value)
else:
  print("Failed to get the data")
```

```python
# get random jokes
def jokes():
  response = requests.get('https://api.chucknorris.io/jokes/random')
  if response.status_code == 200:
    joke_data = response.json()
    return joke_data['value']
  else:
    return "Failed to get the data"
print(jokes())
```

```python
# POST Request
url = "https://jsonplaceholder.typicode.com/posts"
data = {
    "userId": 11,
    "id": 101,
    "title": "Meta Python",
    "body": "Python is a great language"
}
response = requests.post(url, json=data)
if response.status_code == 201:
  post_data = response.json()
  post_id = post_data['id']
  print("Post ID:", post_id)
else:
  print("Failed to post the data")
```

```python
# random password generator
import random
import string
def radnom_password_generator(num_characters):
  characters = string.ascii_letters + string.digits + string.punctuation
  password = ''.join(random.choice(characters)
  for _ in range(num_characters))
  return password

num_characters = int(input("Enter the length of password as u want: "))
password = radnom_password_generator(num_characters)
print(password)
```

Random Quote Engine: Design a function that selects and displays a random quote from a predefined list. https://api.quotable.io/random https://api.chucknorris.io/jokes/random

```python
import requests
import random
def get_random_quote():
  api_endpoints = ['https://api.quotable.io/random', 'https://api.chucknorris.io/jokes/random']
  endpoint = random.choice(api_endpoints)
  response = requests.get(endpoint)
  if endpoint == 'https://api.quotable.io/random':
    data = response.json()
    quote = data['content']
    author = data['author']
    print(f"Quote: \"{quote}\", - {author}")
  elif endpoint == 'https://api.chucknorris.io/jokes/random':
    data = response.json()
    joke = data['value']
    print(f"Joke: \"{joke}\"")
get_random_quote()
```
