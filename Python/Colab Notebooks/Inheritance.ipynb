```python
class person: # parent calss or base class
  def __init__(self,name,phonenumber):
    self.name = name
    self.phonenumber = phonenumber

  def task(self):
    print(self.name, self.phonenumber)

class person2(person): # child class or derived class
  def task2(self):
    print(self.name, self.phonenumber)

object1 = person2()
print(object1.task2())
```

```python
class grandfather: # level 1
  pass
class father(grandfather): # level 2
  pass
class child(father): # (multilevel inheritance) level 3
  pass
```

```python
# multiple inheritance
class a: # base
  pass
class b: # base
  pass
class c(a,b): # child class is derived from a and b
  pass
```

```python
# hierarchical inheritance
class a: # single parent class
  pass
class b(a): # child class b is derived from a
  pass
class c(a): # child class c is derived from a
  pass
```

```python
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        print(self.num1 + self.num2)
    def sub(self):
        print(self.num1 - self.num2)

    def mul(self):
        print(self.num1 * self.num2)

    def div(self):
        if self.num2 == 0:
            print("Error: Cannot divide by zero!")
        else:
            print(self.num1 / self.num2)
class Calc(Calculator):
    pass
object1 = Calc(10, 5)
object1.add()
object1.sub()
object1.mul()
object1.div()
```

```python
# create a sum of 100 numbers using single inheritance
class sum:
  def __init__(self, n, sum):
    self.n = n
    self.sum = 0
    for i in range(1, n + 1):
      self.sum += i
    print("The sum of first", n, "natural numbers is:", self.sum)
class sum2(sum):
  pass
object1 = sum2(100, 0)
```
