```python
class Person: #parent class or base class
  def __init__(self,name,phonenumber):
    self.name = name
    self.phonenumber = phonenumber

  def task(self):
    print(self.name,self.phonenumber)

class Person2(Person): #child class or derived class
      def task2(self):
        print(self.name,self.phonenumber)
object1 = Person2("steve",7000088877)
object1.task2()
```

```python
class a:
  pass

class b(a): #single inheritance
  pass
```

```python
class grandfather:
  pass
class father(grandfather):
  pass
class son(father):
  pass #multilevel inheritance
```

```python
#multiple inheritance
class a: #base
  pass
class b:#base
  pass
class c(a,b): #child class is derived from a and b,single c child class
  pass
```

```python
#hierarchical inheritance

class a: # single parent class
  pass

class b(a): #child class b
  pass
class c(a): #child class c
  pass
```
