```python
# create a dictionary with all the operations
dict = {"+": "a+b",
        "-": "a-b",
        "*": "a*b",
        "/": "a/b",
        "%": "a%b",
        "**": "a**b",
        "//": "a//b",
        }
y = dict["+"]
print(y)
```

```python
x = dict.get("+")
print(x)
```

```python
x = dict.keys()
print(x)
```

```python
x = dict.values()
print(x)
```

```python
dict["^"] = "a^b"
print(dict)
```

```python
dict.update({"+": "b+a"})
print(dict)
```

```python
x = dict.setdefault("**", "b**a")
print(x)
```

```python
dict.pop("^") # it remove the item fomr dictionary with the specfic key name
print(dict)
# OR
del dict["-"]
print(dict)
```

```python
dict.popitem() # it will remove the last inserted item form the dictionary
print(dict)
```

```python
for x, y in dict.items(): # loop for print the keys with values
  print(x, y)
```

```python
# it will print the keys of dictionary
for x in dict.keys():
  print(x)
```

```python
# it will print the values
for x in dict.values():
  print(x)
```

```python
# it will aslo print the values of dictionary
for x in dict:
  print(dict[x])
```

```python
# it will also print all the keys
for x in dict:
  print(x)
```

```python
# Union
set1 = {"apple", "banana", "red", "blue", "cherry"}
set2 = {1, 5, 7, 9, 3}
set3 = set1.union(set2)
print(set3)
```

```python
# Union or "|"
set1 = {"apple", "banana", "red", "blue", "cherry"}
set2 = {1, 5, 7, 9, 3}
set3 = set1 | set2
print(set3)
```

```python
# join a set nand a tuple
set1 = {"apple", "banana", "red", "blue", "cherry"}
tuple1 = ("apple", "banana", "red", "blue", "cherry")
set3 = set1.union(tuple1)
print(set3)
```
